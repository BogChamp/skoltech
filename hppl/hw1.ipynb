```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bifurcation_map(x, r):
    return r * x * (1 - x)
```

## 1

```python
n_steps = 100 
x0 = 0.2
r = 1
xs = [0.0 for _ in range(n_steps + 1)]
xs[0] = x0

for i in range(n_steps):
    xs[i + 1] = bifurcation_map(xs[i], r)
```

```python
plt.plot(xs, marker='.')
plt.title('Bifurcation of x')
plt.xlabel('step')
plt.ylabel('x')
plt.show()
```

```python
xs = [0.0 for _ in range(n_steps + 1)]
xs[0] = x0

for i in range(n_steps):
    xs[i + 1] = bifurcation_map(xs[i], 3)

plt.plot(xs, marker='.')
plt.title('Bifurcation of x')
plt.xlabel('step')
plt.ylabel('x')
plt.show()
```

```python
xs = [0.0 for _ in range(n_steps + 1)]
xs[0] = 0.5

for i in range(n_steps):
    xs[i + 1] = bifurcation_map(xs[i], 2)

plt.title('Bifurcation of x')
plt.xlabel('step')
plt.ylabel('x')
plt.plot(xs, marker='.')
plt.show()
```

## 2

```python
rs = np.linspace(2, 5, 10000)
n = 2000
m = 200
last_ms = [None] * m 

x = np.array([0.1 for _ in rs])

for i in range(n):
    x = bifurcation_map(x, rs)

for i in range(m):
    x = bifurcation_map(x, rs)
    last_ms[i] = bifurcation_map(x, rs)
```

## 3

```python
for i in range(m):
    plt.plot(rs, last_ms[i], ', g')

plt.ylabel('x')
plt.xlabel('r')
plt.show()
```
