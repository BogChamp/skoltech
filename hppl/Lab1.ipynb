````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
even_list = [i for i in range(100) if i % 2 == 0]
````

````python
list(reversed(even_list))
````

````python
import csv

with open('even.csv', 'w') as f:
    write = csv.writer(f)   
    write.writerow(even_list)


with open('even.csv', 'r') as f:
    reader = csv.reader(f)   
    even_list = list(reader)[0]
    even_list = list(map(int, even_list))
````

````python
with open('bin.bin', 'wb') as f:
    f.write(bytearray(even_list))

with open('bin.bin', 'rb') as f:
    even_list = list(f.read())
````

````python
import pickle

with open('even.pkl', 'wb') as f:
    pickle.dump(even_list, f)

with open('even.pkl', 'rb') as f:
    even_list = pickle.load(f)
````

````python
plt.plot(x, y)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xticks(np.arange(min(x), max(x)+1, 10), fontsize=14)
plt.savefig('graph.png', dpi=150, format='png')
plt.savefig('graph.pdf', dpi=150, format='pdf')
plt.show()
````

* На отдельном графике нарисовать график матрицы, где все элементы вне круга с центром в (0.5м,0.5м) и радиусом (0.25м) зануляются.

````python
import numpy as np
import matplotlib.pyplot as plt

class MyMatrix:
    def __init__(self, n=10):
        self.matrix = np.random.rand(n,n).astype(np.float32)
        self.n = n
        self.x = np.linspace(0, 1, self.n)
        self.y = np.linspace(0, 1, self.n)

    def max_value(self):
        print(np.amax(self.matrix))
    def plot_matrix(self):
        plt.imshow(self.matrix, origin='lower', aspect='auto')
        ind = np.unravel_index(np.argmax(self.matrix, axis=None), self.matrix.shape)
        plt.plot(ind[1], ind[0], 'ro')
        plt.colorbar()
    def plot_matrix_meters(self):
        plt.imshow(self.matrix, origin='lower', aspect='auto', extent=[0,1,0,1])
        plt.plot(0.5, 0.2, 'ro')

    def plot_matrix_with_mask(self):
        x, y = np.meshgrid(self.x, self.y)
        self.matrix[(x-0.5)**2+(y-0.5)**2>0.25**2] = 0
        plt.imshow(self.matrix, origin='lower', aspect='auto', extent=[0,1,0,1])
````

````python
a = MyMatrix(50)
a.plot_matrix()
````

````python
a.plot_matrix_meters()
````

````python
a.plot_matrix_with_mask()
````

````python
grid_min = -20 * 2 * np.pi
grid_max = 20 * 2 * np.pi
duration = 5 * np.pi

t = np.linspace(-20 * 2 * np.pi, 20 * 2 * np.pi, 1000)
y = np.exp(-t**2 / 2 / duration**2) * np.sin(t)

plt.plot(t / 2/ np.pi,y)
plt.xlim(0,10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

````python
sp = np.fft.fft(y)
w = np.fft.fftfreq(len(y), d=(t[1]-t[0]) / (np.pi * 2))

dt = t[1]-t[0]
dw = 1 / len(y) / dt


plt.plot(w, abs(sp))
plt.xlim(0,4)
````
