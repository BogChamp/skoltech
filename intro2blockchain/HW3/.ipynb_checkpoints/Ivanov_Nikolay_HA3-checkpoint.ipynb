**(2 points)** From Task_1 connect to the Ethereum network and provide codes to interact with the blockchain:
 * Checking block height at any time.
 * Getting information about a block.
 * Get balance information of your address_1.
 * From Task 3.3. Convert the balance to ether.
(Hint: https://web3js.readthedocs.io/en/v1.2.4/ or https://web3py.readthedocs.io/)

````python
network = 'goerli'
provider_url = 'https://goerli.infura.io/v3/1f487ff1f4a9446b95206a2bbffb8c4a'

web3 = Web3(Web3.HTTPProvider(provider_url))

# check block height
block_height = web3.eth.block_number
print(f"Block height on {network} testnet: {block_height}")

# get block information
block_number = 8652886
block = web3.eth.get_block(block_number)
print(f"Block number {block_number} on {network} testnet:")
print(block)

# get balance of address_1
address_1 = address_1
balance_wei_1 = web3.eth.get_balance(address_1)
balance_ether_1 = web3.fromWei(balance_wei_1, 'ether')

adress_2 = address_2
balance_wei_2 = web3.eth.get_balance(address_2)
balance_ether_2 = web3.fromWei(balance_wei_2, 'ether')

print(f"Balance of {address_1} on {network} testnet: {balance_wei_1} wei")
print(f"Balance of {address_1} on {network} testnet: {balance_ether_1} ether")
print(f"Balance of {address_2} on {network} testnet: {balance_wei_2} wei")
print(f"Balance of {address_2} on {network} testnet: {balance_ether_2} ether")
````

**(1 point)** Send some Eth test coins from address_1 to address_2 and provide Transaction_ID.

````python
balance_before_1 = web3.eth.get_balance(address_1)
print(f"Balance of address_1 before Transaction: {balance_before_1} wei")
balance_before_2 = web3.eth.get_balance(address_2)
print(f"Balance of address_2 before Transaction: {balance_before_2} wei")

nonce = web3.eth.get_transaction_count(address_1)
# build the transaction
tx = {
    'nonce':nonce,
    'from': address_1,
    'to': address_2,
    'value': web3.toWei(0.01, 'ether'),  # send 0.05 ETH
    'gas': 30000,
    'gasPrice': web3.toWei('50', 'gwei')
}

print(f'Sending ETH from {address_1} to {address_2}')

# sign and send the transaction
signed_tx = web3.eth.account.sign_transaction(tx, private_key=private_key_1)
tx_hash = web3.eth.send_raw_transaction(signed_tx.rawTransaction)

# get the transaction receipt and ID
tx_receipt = web3.eth.wait_for_transaction_receipt(tx_hash)
tx_id = tx_receipt.transactionHash.hex()
print(f"Transaction ID: {tx_id}")
````

````python
balance_after_1 = web3.eth.get_balance(address_1)
print(f"Balance of address_1 before Transaction: {balance_after_1} wei")
balance_after_2 = web3.eth.get_balance(address_2)
print(f"Balance of address_2 before Transaction: {balance_after_2} wei")
````
