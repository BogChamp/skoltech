Now let's estimate $|R(N)|$:$$|R(N)| = |\sum\limits_{n=N+1}^{\infty}\frac{\partial^n f(x_i,0)}{\partial y^n}\frac{y_j^n}{n!}| \leq \sum\limits_{n=N+1}^{\infty}|\frac{\partial^n f(x_i,0)}{\partial y^n}|\frac{y_j^n}{n!} \leq \sum\limits_{n=N+1}^{\infty} M \frac{y_j^n}{n!} \leq M \sum\limits_{n=N+1}^{\infty}\frac{1}{n!}$$ since $y_j \leq 1$ $$\sum\limits_{n=N+1}^{\infty}\frac{1}{n!} = \frac{1}{(N+1)!}(1+\sum\limits_{n=2}^{\infty}\frac{1}{\prod_{k=2}^n(N+k)}) \leq \frac{1}{(N+1)!}\sum\limits_{n=0}^{\infty}(\frac{1}{N+2})^n = \frac{1}{(N+1)!}\frac{N+2}{N+1} \leq \frac{2}{(N+1)!}$$
And since $|R(N)| \leq \epsilon$ let's take such N that $\frac{2M}{(N+1)!} \leq \epsilon \to (N+1)! \geq \frac{2M}{\epsilon}$. Let's use upper bound on n!: $n! \leq \frac{n^{n+1}}{e^{n-1}}$. Thus $$\frac{2M}{\epsilon} \leq (N+1)! \leq \frac{(N+1)^{N+2}}{e^{N}} \leq (N+1)^{N+2} \to \ln(\frac{2M}{\epsilon}) \leq (N+2)\ln(N+1) \leq (N+2)^2 \to \sqrt{\ln(\frac{2M}{\epsilon})}-2 \leq N$$
Thus we can take $N = \sqrt{\ln(\frac{2M}{\epsilon})}$ and it will lead to $|A-X|_{\infty}\leq \epsilon$. So X can be expressed as sum of $\sqrt{\ln(\frac{2M}{\epsilon})}$ matricies with rank 1, which means that $\epsilon$ rank bounded with $\sqrt{\ln(\frac{2M}{\epsilon})}$ this bound doesn't depend on p,q and choice of points.
+ (6 pts). Extend result from the previous point (a.) to $x, y \in \mathbb{R}^{N}$. Observe how the bound depends on $N$.

In previous section we used such properties that all derivatives bounded at points $(x,0)$ and every number $(y-0)^n$ bounded with some number, more specifically with 1. So for extending results on $R^N$ we need more conditions or limitations.

Firstly, we can just bound domain of definition for $y$ by some area, for example $[-L,L]^N$, and secondly, limit all derivatives $|\frac{\partial^n f(x_1,x_2\ldots,x_N,0,\ldots,0)}{\partial y_1^{\alpha_1}\ldots \partial y_N^{\alpha_N}}| < M$ where $\sum \alpha_i = n$.\
In this situation we have bound on remainder:
$$|R(K)| = |\sum\limits_{n=K+1}^{\infty}\sum\limits_{\sum\alpha_i=n}\frac{\partial^n f(x_{1i},\ldots,x_{Ni},0,\ldots,0)}{\partial y_1^{\alpha_1}\ldots \partial y_N^{\alpha_N}}\frac{y_{1j}^{\alpha_1}y_{2j}^{\alpha_2}\ldots y_{Nj}^{\alpha_N}}{n!}| \leq \sum\limits_{n=K+1}^{\infty}\frac{N^n M L^n}{n!} \leq \frac{M*(NL)^{K+1}}{(K+1)!}(1+\sum\limits_{n=2}^{\infty}\frac{(NL)^{n-1}}{\prod_{k=2}^n(K+k)}) \leq \frac{2M*(NL)^{K+1}}{(K+1)!}\leq \epsilon$$
Last two inequalities holds due to proven above and assumption that we take such big K that $NL/(K+2) < 1$. First holds due to modulus unequation and fact that sum of multinomial coefficients equal to $N^n$, where N - num of elements in sum and n - power of sum. Now:
$$\frac{(NL)^{K+1}}{(K+1)!}\leq \frac{\epsilon}{2M} \to \frac{2M}{\epsilon} \leq \frac{(K+1)!}{(NL)^{K+1}} \leq (\frac{K+1}{NL})^{K+1} \to \ln(\frac{2M}{\epsilon}) \leq (K+1)\frac{K+1}{NL} \to \sqrt{NL\ln(\frac{2M}{\epsilon})}\leq K+1$$
Thus, epsilon rank of matrix A bounded with $\sqrt{NL\ln(\frac{2M}{\epsilon})}-1$. Also, we can extend result on every compact set in $R^N$, and thus L - it's a radius of sphere, which contains whole given domain of definition of function.

```python
def find_similar_molecules(molecules):
    bc = molecules[:,1,:] - molecules[:,2,:]
    cd = molecules[:,2,:] - molecules[:,3,:]
    de = molecules[:,3,:] - molecules[:,4,:]
    eb = molecules[:,4,:] - molecules[:,1,:]
    vecs = np.stack((bc,cd,de, eb),axis=1)
    norms = np.round(np.linalg.norm(vecs, axis=2), 6)
    classes = defaultdict(list)
    for i,v in enumerate(norms):
        classes[tuple(v)].append(molecules[i])
    return classes
```

+ (7 pts). Learn about [Kabsch algorithm](https://en.wikipedia.org/wiki/Kabsch_algorithm). Implement it and run on the database. Produce a list of distinct molecules and visualise them.

```python
def kabsch_algo(molecule1, molecule2):
    h = jnp.matmul(molecule1.T, molecule2)
    u,s,vt = jnp.linalg.svd(h)
    d = jnp.linalg.det(jnp.matmul(vt.T, u.T))
    e = jnp.array([[1,0],[0,d]])
    r = jnp.matmul(jnp.matmul(vt.T, e), u.T)
    return r
```

```python
def clasterize(molecules):
    key = False
    classes = defaultdict(list)
    j = -1
    for i in molecules:
        for j,v in classes.items():
            r = kabsch_algo(i, v[0])
            if not np.round(jnp.linalg.norm(i-v[0]@r),5):
                classes[j].append(i)
                key = True
                break
        if not key:
            classes[j+1].append(i)
        key = False
    return classes
```

```python
kabsch_classes = clasterize(molecules)
```

# Problem 4 (Disentangled representation) (12 pts)

## Brief intro

TheÂ mainÂ componentÂ ofÂ [GAN](https://pytorch.org/tutorials/beginner/dcgan_faces_tutorial.html)Â isÂ GeneratorÂ network.Â GeneratorÂ isÂ suchÂ modelÂ thatÂ takesÂ asÂ inputÂ randomÂ vectorÂ andÂ producesÂ anÂ image.Â InÂ thisÂ task,Â weÂ willÂ workÂ withÂ theÂ alreadyÂ pre-trainedÂ generatorÂ $G$.Â HereÂ areÂ aÂ fewÂ thingsÂ youÂ needÂ toÂ knowÂ aboutÂ thisÂ model:



1.Â Â Â **ModelÂ input:**Â batchÂ ofÂ randomÂ vectorsÂ $z\simÂ N(0,1)$Â withÂ normallyÂ distributedÂ valuesÂ ofÂ sizeÂ **(Batch_Size,100)**.Â Here,Â Batch_SizeÂ isÂ a number ofÂ imagesÂ weÂ wantÂ toÂ generate
2.Â Â Â **ModelÂ output:**Â $\mathrm{images}Â =Â G(z)$Â -Â batchÂ ofÂ imagesÂ withÂ sizeÂ **(Batch_Size,3,64,64)**.Â HereÂ **3**Â -Â definesÂ numberÂ ofÂ imageÂ channelsÂ (3Â forÂ RGBÂ images),Â **64,Â 64**Â -Â widthÂ andÂ height.
3.Â Â Â **ModelÂ structure:**Â generatorÂ isÂ aÂ setÂ ofÂ sequentiallyÂ apppliedÂ functions.Â EachÂ functionÂ inputÂ isÂ theÂ outputÂ ofÂ the previousÂ one.Â InÂ otherÂ words,Â $G(z)Â =Â f_N(f_{N-1}...f_1(f_0(z)))$.Â $N$Â -Â definesÂ numberÂ ofÂ functionsÂ inÂ generator.



### Disentanglement

TheÂ goodÂ propertyÂ ofÂ GANÂ isÂ thatÂ generatorÂ inputÂ spaceÂ hasÂ aÂ setÂ ofÂ meaningfulÂ directionsÂ -Â movingÂ inputÂ vectorÂ alongÂ thisÂ directionsÂ produceÂ interpretableÂ changesÂ inÂ image.Â ForÂ example,Â weÂ canÂ changeÂ aÂ Pose,Â Age,Â ExpressionÂ orÂ evenÂ addÂ eyeglasses.Â YouÂ canÂ readÂ moreÂ aboutÂ itÂ onÂ [resource](https://genforce.github.io/interfacegan/).Â TheÂ mainÂ questionÂ isÂ howÂ toÂ findÂ suchÂ goodÂ directions.Â ThisÂ problemÂ isÂ calledÂ disentanglement.

![](https://github.com/genforce/interfacegan/blob/master/docs/assets/teaser.jpg?raw=true)

## Problem discription 

WeÂ areÂ goingÂ toÂ findÂ meaningfulÂ directionsÂ usingÂ SVDÂ appliedÂ overÂ theÂ Jacobian matrixÂ ofÂ someÂ hiddenÂ outputÂ withÂ respectÂ toÂ modelÂ input. **i-th**Â hiddenÂ outputÂ ofÂ generatorÂ modelÂ is defined by formulaÂ $G_{i}(z)Â =Â f_i(f_{i-1}...f_1(f_0(z))))$,Â whereÂ $iÂ <Â N$.Â So,Â theÂ taskÂ couldÂ beÂ splittedÂ onÂ fourÂ mainÂ parts:

1.Â **(4Â pts)**Â Â Â WriteÂ theÂ functionÂ **Get_Hidden_Output**,Â whichÂ reproducesÂ formulaÂ $G_{i}(z)Â =Â f_i(f_{i-1}...f_1(f_0(z))))$
2.Â **(3Â pts)**Â Â Â CalculateÂ jacobian matrixÂ $J$Â ofÂ generatorÂ $G$Â forÂ severalÂ **i**Â values.Â ForÂ eachÂ ofÂ them:

*Â Â Â CalculateÂ theÂ jacobian matrixÂ withÂ respectÂ toÂ $z$:Â $\frac{\partialÂ G(z)}{\partialÂ z}$
*Â Â Â ReshapeÂ jacobian matrixÂ toÂ haveÂ theÂ sizeÂ **(Batch_Size * Hidden_Output_Size,100)**

3.Â **(3Â pts)**Â Â ApplyÂ SVDÂ toÂ jacobianÂ matrixÂ ($JÂ =Â UÂ @Â SÂ @Â V.T$)Â fromÂ previousÂ step

4.Â **(2Â pts)**Â Â TakeÂ severalÂ directionsÂ fromÂ matrixÂ $V$Â (particularÂ columns),Â andÂ runÂ visualizationÂ codeÂ (alreadyÂ writtenÂ forÂ you).Â LeaveÂ aÂ commentÂ onÂ whatÂ youÂ gotÂ (howÂ foundÂ directionsÂ affectÂ image)

```python
class Generator(nn.Module):
    def __init__(self, ngpu):
        super(Generator, self).__init__()
        self.ngpu = ngpu
        self.main = nn.Sequential(
            # input is Z, going into a convolution
            nn.ConvTranspose2d( nz, ngf * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(ngf * 8),
            nn.ReLU(True),
            # state size. (ngf*8) x 4 x 4
            nn.ConvTranspose2d(ngf * 8, ngf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 4),
            nn.ReLU(True),
            # state size. (ngf*4) x 8 x 8
            nn.ConvTranspose2d( ngf * 4, ngf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 2),
            nn.ReLU(True),
            # state size. (ngf*2) x 16 x 16
            nn.ConvTranspose2d( ngf * 2, ngf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf),
            nn.ReLU(True),
            # state size. (ngf) x 32 x 32
            nn.ConvTranspose2d( ngf, nc, 4, 2, 1, bias=False),
            nn.Tanh()
            # state size. (nc) x 64 x 64
        )
    def get_function_output(self, inp, index=0):
      if index == 0:
        inp = inp[:,:,None,None]
      return self.main[index](inp)
    def get_number_of_layers(self):
      return len(self.main)

    def forward(self, input):
        return self.main(input[:,:,None,None])
```

```python
def Calc_Gradient(z, index):
  grads = vmap(jacfwd(lambda z: Get_Hidden_Output(z, index=index)))(z.unsqueeze(1))
  return grads
```
